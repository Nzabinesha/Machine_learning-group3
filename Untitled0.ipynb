```python
import numpy as np
import pandas as pd

# ---------------------------------------------------------------
# 1. LOAD + POOL THE DATA
# ---------------------------------------------------------------
df = pd.read_csv("GaltonFamilies.csv")
fathers = df.drop_duplicates(subset="family")["father"].values
children = df["childHeight"].values
x = np.concatenate([fathers, children])
n = len(x)

print(f"Pooled dataset size: n = {n}")
print(f"(True) father mean={fathers.mean():.3f}, children mean={children.mean():.3f}")
print(f"Global mean of pooled data: {x.mean():.3f}\n")

# ---------------------------------------------------------------
# 2. EM FUNCTIONS
# ---------------------------------------------------------------
def gaussian_pdf(x, mu, sigma2):
    return (1.0 / np.sqrt(2 * np.pi * sigma2)) * np.exp(-((x - mu)**2) / (2 * sigma2))

def e_step(x, mu1, mu2, sigma1_sq, sigma2_sq, pi1, pi2):
    w1 = pi1 * gaussian_pdf(x, mu1, sigma1_sq)
    w2 = pi2 * gaussian_pdf(x, mu2, sigma2_sq)
    total = w1 + w2
    return w1 / total, w2 / total

def m_step(x, gamma1, gamma2):
    N1, N2 = gamma1.sum(), gamma2.sum()
    mu1, mu2 = np.sum(gamma1 * x) / N1, np.sum(gamma2 * x) / N2
    s1 = np.sum(gamma1 * (x - mu1)**2) / N1
    s2 = np.sum(gamma2 * (x - mu2)**2) / N2
    return mu1, mu2, s1, s2, N1/len(x), N2/len(x)

def log_likelihood(x, mu1, mu2, sigma1_sq, sigma2_sq, pi1, pi2):
    dens = (pi1 * gaussian_pdf(x, mu1, sigma1_sq) + pi2 * gaussian_pdf(x, mu2, sigma2_sq))
    return np.sum(np.log(dens))

# ---------------------------------------------------------------
# 3. EXECUTION
# ---------------------------------------------------------------
x_sorted = np.sort(x)
mu1, mu2 = x_sorted[:n//2].mean(), x_sorted[n//2:].mean()
sigma1_sq = sigma2_sq = x.var()
pi1 = pi2 = 0.5

table_rows = [(0, mu1, mu2, sigma1_sq, sigma2_sq, pi1, pi2, log_likelihood(x, mu1, mu2, sigma1_sq, sigma2_sq, pi1, pi2))]

for it in range(1, 100):
    g1, g2 = e_step(x, mu1, mu2, sigma1_sq, sigma2_sq, pi1, pi2)
    mu1_n, mu2_n, s1_n, s2_n, p1_n, p2_n = m_step(x, g1, g2)
    ll = log_likelihood(x, mu1_n, mu2_n, s1_n, s2_n, p1_n, p2_n)
    if it <= 2: table_rows.append((it, mu1_n, mu2_n, s1_n, s2_n, p1_n, p2_n, ll))
    if abs(ll - table_rows[-1][-1]) < 1e-4 and it > 2: break
    mu1, mu2, sigma1_sq, sigma2_sq, pi1, pi2 = mu1_n, mu2_n, s1_n, s2_n, p1_n, p2_n

print(f"{'Iter':<5}{'mu1':<10}{'mu2':<10}{'sigma1^2':<12}{'sigma2^2':<12}{'pi1':<8}{'pi2':<8}{'Log-Lik':<12}")
for r in table_rows: print(f"{r[0]:<5}{r[1]:<10.4f}{r[2]:<10.4f}{r[3]:<12.4f}{r[4]:<12.4f}{r[5]:<8.4f}{r[6]:<8.4f}{r[7]:<12.4f}")

# ---------------------------------------------------------------
# 4. BASELINE & CLASSIFICATION (Person B Presentation Tasks)
# ---------------------------------------------------------------
test_h = 71.0
g1, g2 = e_step(np.array([test_h]), mu1, mu2, sigma1_sq, sigma2_sq, pi1, pi2)
p_child, p_pro = (g1[0], g2[0]) if mu1 < mu2 else (g2[0], g1[0])
print(f"\nTest height = {test_h} in\nP(Child) = {p_child:.4f}\nP(Pro)   = {p_pro:.4f}")

global_mean = x.mean()
pile_low, pile_high = x[x <= global_mean], x[x > global_mean]
print(f"\n--- Naive mean-split baseline ---\nGlobal mean cutoff: {global_mean:.3f}")
print(f"Pile 1 (<= mean): n={len(pile_low)}, mean={pile_low.mean():.3f}")
print(f"Pile 2 (>  mean): n={len(pile_high)}, mean={pile_high.mean():.3f}")

lower_mu = min(mu1, mu2)
overlap = x[(x > lower_mu - 3) & (x < lower_mu + 3)]
print(f"Points within +/-3in of children component mean: {len(overlap)} / {n} ({100*len(overlap)/n:.1f}%)")
```
